```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
### read data
df = pd.read_csv('simulated_data.csv',sep=';')
```

```python
### show single product
df[df.PRODUCT_ID==20]['SALES'].hist(bins=100)
```

```python
#plot time series of product
df[df.PRODUCT_ID==20].plot(x='DATE',y='SALES')
```

```python
# plot sum of all SALES 
df.groupby('DATE').sum()['SALES'].plot()
```

```python
df.groupby('PRODUCT_ID').mean()['SALES'].plot(kind='bar')
```
