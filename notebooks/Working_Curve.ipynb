```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
### read data
df = pd.read_csv('result_rmean.csv',sep=';')
```

```python
### Plot Curves
df = df.sort('OOS')
df.set_index('OOS',inplace=True)

fig, ax = plt.subplots()

df.groupby('VERSION_ID')['SURPLUS'].plot(labels='VERSION_ID', style='o-',ax=ax)

for k, v in df.iterrows():
    ax.annotate(str(v['QUANTILE']),(k,v['SURPLUS']))

plt.ylabel('EXCESS RATE')
plt.xlabel('OOS RATE')
plt.legend()

fig.canvas.draw()
```
